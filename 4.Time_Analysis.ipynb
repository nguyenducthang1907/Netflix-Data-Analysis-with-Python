```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns
df = pd.read_csv(r"D:\Application\Porfolio\Poject With Python\Netflix Dataset\netflix_cleaned.csv")
df.dtypes
```

```python
df.head()
```

```python
# df['date_added'] = pd.to_datetime(df['date_added'])
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month_name()
content_per_year = df['added_year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=content_per_year)
plt.title('Total Content Added to Netflix Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.grid(True)
plt.show()
```
